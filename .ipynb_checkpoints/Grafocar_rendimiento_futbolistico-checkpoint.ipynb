# Graficar rendimiento futbolÃ­stico

importar las librerias usadas

```python
import re # modulo para usar expresiones regulares
from pandas import DataFrame # modulo para almacenar los datos limpios
import matplotlib.pyplot as plt # modulo para graficar los datos
```

declarar las constantes usadas

```python
RUTA_ARCHIVO_CSV = "./results.csv" # Fuente: Curso de Expresiones Regulares en Platzi (https://static.platzi.com/media/public/uploads/results_6aeb6252-c531-449d-bf29-e11193358b8c.csv)
EQUIPO_FUTBOL = "Mexico" # equipo seleccionado para medir su rendimiento, se debe escribir el nombre como esta en el CSV
STR_GANADOS = "Ganados" # cadena para contabilizar los partidos ganados
STR_EMPATADOS = "Empatados"# cadena para contabilizar los partidos empatados
STR_PERDIDOS = "Perdidos" # cadena para contabilizar los partidos perdidos
STR_TITULO_GRAFICO = "Rendimiento del equipo: " # titulo del grafico
STR_Y = "No. de juegos" # titulo de la variable y
STR_X = "Temporada" # titulo de la variable x
TIPO_GRAFICO = "bar" # tipo de grafico
```

definir funcion para conocer el resultado del partido

```python
# cadenaEquipo, es un string que trae el nombre del equipo local y el visitante, separados por una coma
# ejemplo: MExico,Uruguay
# goles, es una lista que tiene en la posicion 0 los goles anotados por el equipo local y en la posicion 1 los goles del equipo visitante
# ejemplo: [2,2]
def getResultadoPartido(cadenaEquipos="", goles=[]):
    if len(cadenaEquipos) > 0:
        equipos = cadenaEquipos.split(",") # se convierte la cadena en uns lista
        indexEquipoParaMedir = equipos.index(EQUIPO_FUTBOL) # obtener el indice del EQUIPO_FUTBOL
        indexOtroEquipo = 1 if indexEquipoParaMedir == 0 else 0 # obtener el indice del equipo que enfrento el EQUIPO_FUTBOL
        if (goles[indexEquipoParaMedir] > goles[indexOtroEquipo]):
            return STR_GANADOS
        elif (goles[indexEquipoParaMedir] == goles[indexOtroEquipo]):
            return STR_EMPATADOS
        else:
            return STR_PERDIDOS
```

definir las variables del programa

```python
aniosJugados = {} # diccionario que almacena cada anio jugado por EQUIPO_FUTBOL con el no. de ganados, emptados y perdidos
archivo 	= open(RUTA_ARCHIVO_CSV, "r") # archivo CSV cargado en memoria
patron = re.compile(r'^([\d]{4,4})\-\d{2,2}\-\d{2,2},(.*'+EQUIPO_FUTBOL+'.*),(\d+),(\d+).*$') # expresion regular para obtener todos los partidos del EQUIPO_FUTBOL 
```

iterar el archivo csv para obtener los datos del equipo seleccionado con la ayuda de expresiones regulares

```python
for linea in archivo:
    resultado = re.match(patron, linea)
    if resultado:
        anio = int(resultado.group(1)) # ejemplo: 2020
        cadenaEquipos = str(resultado.group(2)) # ejemplo: Mexico,Ururguay
        marcadorGoles = [resultado.group(3), resultado.group(4)] # ejemplo: [2,2]
        resultadoPartido = getResultadoPartido(cadenaEquipos, marcadorGoles) # ejemplo: "Empatados"
        # contabilizar los juegos en ganados, emptados y perdidos
        if anio not in aniosJugados:
            aniosJugados[anio] = {STR_GANADOS:0, STR_EMPATADOS:0, STR_PERDIDOS:0}
            aniosJugados[anio][resultadoPartido] = aniosJugados[anio][resultadoPartido] + 1
        else:
            aniosJugados[anio][resultadoPartido] = aniosJugados[anio][resultadoPartido] + 1
archivo.close()
```

crear dataframe con los juegos encontrados y ordenarlos por el anio

```python
aniosJugadosDf = DataFrame(aniosJugados).T
aniosJugadosDf = aniosJugadosDf.sort_index()
```

configurar grafico

```python
ax = aniosJugadosDf.plot(use_index=True, y=[STR_GANADOS, STR_EMPATADOS, STR_PERDIDOS], 
                    title = STR_TITULO_GRAFICO+EQUIPO_FUTBOL,
                    grid = True,
                    kind = TIPO_GRAFICO)
ax.set_xlabel(STR_X)
ax.set_ylabel(STR_Y) 
```

mostrar grafico

```python
plt.show()
```
